```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
for i in range(1):
        webD.find_element_by_xpath('//*[@id="footer-pager"]/div/div/div[2]/div[1]/a').click()
        webD.implicitly_wait(5000)
```

```python
list_of_links=[]
elements=webD.find_elements_by_css_selector('.product-page')
for el in elements:
    skirts=el.find_elements_by_css_selector('.product-url.js-click-product.js-category-product__link')
    for element in skirts:
        list_of_links.append(element.get_property('href'))    
```

```python
len(list_of_links)
```

```python
list_of_links[0]
```

```python
from tqdm import tqdm
alldetails=[]
for i in tqdm(list_of_links):
    webD.get(i)
    name=webD.find_element_by_xpath('//*[@id="product-right"]/div[1]/h1').text
    old_price=webD.find_element_by_xpath('//*[@id="product-right"]/div[2]/div/p[1]/span/span').text
    new_price=webD.find_element_by_xpath('//*[@id="product-right"]/div[2]/div/p[2]/span[1]/span').text
    description=webD.find_element_by_xpath('//*[@id="product-right"]/div[11]').text
    temp={'name':name,'old_price':old_price,'new_price':new_price,'description':description,'link':i}
    alldetails.append(temp)
```

```python
import pandas as pd
pd.DataFrame(alldetails)
```
